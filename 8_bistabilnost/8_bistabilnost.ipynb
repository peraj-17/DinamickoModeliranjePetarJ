### **2. Ako pođemo od nagiba $k = 0.8$, pri čemu se $k$ postepeno smanjuje, u jednom trenutku će prava linija postati tangenta na sigmoidnu krivu, dakle dodirivaće je u jednoj tački. Nacrtajte ovaj slučaj na računaru.**

U ovom slučaju važiće da za datu vrednost x, dve linije se seku ($f(x) = kx$) i imaju isti nagib ($f'(x) = k$). 

\begin{equation}
    f(x) = \frac{x^2}{1 + x^2}
\end{equation}

\begin{equation}
    f'(x) = \frac{2x}{(1 + x^2)^2}
\end{equation}

Prvi uslov:

\begin{equation}
    kx = \frac{x^2}{1 + x^2} \Rightarrow k = \frac{x}{1 + x^2}
\end{equation}

Drugi uslov:

\begin{equation}
    kx = \frac{2x}{(1 + x^2)^2} \Rightarrow k = \frac{2}{(1 + x^2)^2} \Rightarrow 1 + x^2 = 2 \Rightarrow x = 1
\end{equation}

Ako uvrstimo $x = 1$ u prvu jednačinu, dobijamo:

\begin{equation}
    k_c = \frac{1}{1 + 1} = 0.5
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Jednačina sigmoide
def f(x):
    return x**2 / (1 + x**2)
```

```python
# Parametri
x = np.linspace(0, 4, 1000)
kc_values = [0.8, 0.6, 0.5, 0.4, 0.3]
```

```python
# Grafik
plt.figure(figsize=(8,6))
plt.plot(x, f(x), label=r"$y=\frac{x^2}{1+x^2}$")
for kc in kc_values:
    plt.plot(x, kc*x, label=f"y = {kc}x")
plt.scatter([1], [0.5], color="red", zorder=3)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Stacionarna stanja za različite vrednosti $k_c$")
plt.ylim(0, 1.05)
plt.xlim(0, 4)
plt.legend()
plt.tight_layout()
plt.show()
```

### **9. Ovakva bifurkacija se naziva sedlo-čvor. Motivaciju za taj naziv nije teško sagledati iz prethodnih zadataka u vežbi. Naime, videli smo da smanjenjem $k$, odnosno prolaskom kroz $k_c$, nastaju dva nova stacionarna stanja, jedno nestabilno (sedlo) i drugo stabilno (čvor). Suprotno, ako krenemo od manjih vrednosti $k$ (za koje je sistem bistabilan), koje onda povećavamo, u $k_c$ će se dva stacionarna stanja sudariti i međusobno poništiti. Kakva je stabilnost stacionarnog stanja koje se javlja u $k_c$, odnosno kada se desi "sudar"?**

U tački $k = k_c$, nastaje stacionarno stanje koje nije u potpunosti stabilno, ali ni u potpunosti nestabilno. U stvari, ono predstavlja granični slučaj između postojanja para stacionarnih stanja i njihovog potpunog nestanka. 

```python
# Parametri
kc = 0.5

# Vrednosti k za x != 0
k = np.linspace(1e-4, kc, 500)

D = 1 - 4*k**2

x_plus  = (1 + np.sqrt(D)) / (2*k)
x_minus = (1 - np.sqrt(D)) / (2*k)

# Vrednosti k za x = 0
k_all = np.linspace(0, 1.0, 500)
x0 = np.zeros_like(k_all)
```

```python
# Bifurkacioni dijagram
plt.figure(figsize=(8,6))

# x = 0 grana
plt.plot(k_all, x0, label="x = 0")

# x != 0 grane
plt.plot(k, x_minus, label="$x_-$ (donja vrednost)")
plt.plot(k, x_plus,  label="$x_+$ (gornja vrednost)")

plt.scatter([kc], [1.0], zorder=3)
plt.xlabel("k")
plt.ylabel("Stacionarna stanja x")
plt.title("Bifurkacioni dijagram sedlo-čvor")
plt.ylim(-0.1, 10)       
plt.xlim(0, 1.0)
plt.legend()
plt.tight_layout()
plt.show()
```
